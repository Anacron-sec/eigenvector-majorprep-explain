```python
import numpy as np
```

```python
markov_matrix = np.array([[.8, .1],[.2, .9]])

human_zombies = np.array([125, 175])
```

By applying the operation a lot of times we can see the operation converges to the vector [100, 200] which is our eigenvector

```python
for i in range(1000):
    human_zombies = markov_matrix.dot(human_zombies)
    print(human_zombies)
```
